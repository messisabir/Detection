```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("fleur.jpeg")
```

```python
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(image)
```

```python
cv2.waitKey(0)
```

```python
R , G , B = cv2.split(image)
```

```python
_, thres = cv2.threshold(B , 100 , 256 , cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
contour =cv2.Canny(thres , 100 , 200 )
```

```python
plt.subplot(331)
plt.imshow(image)
plt.title("ORIGINAL")
plt.axis("off")

plt.subplot(332)
plt.imshow(R , cmap="gray")
plt.title("image rouge")
plt.axis("off")

plt.subplot(333)
plt.imshow(G , cmap="gray")
plt.title("image vert")
plt.axis("off")

plt.subplot(334)
plt.imshow(B , cmap="gray")
plt.title("image blue")
plt.axis("off")

plt.subplot(335)
plt.imshow(thres, cmap="gray")
plt.title("OTSU")
plt.axis("off")

plt.subplot(336)
plt.imshow(contour , cmap="gray")
plt.title("detection des contour")
plt.axis("off")

plt.show()
```

```python
#exmple sur des image
contour =cv2.Canny(R , 100 , 200 )
plt.subplot(336)
plt.imshow(contour , cmap="gray")
plt.title("detection des contour")
plt.axis("off")

plt.show()
```

```python
#exmple sur des image
contour =cv2.Canny(G , 100 , 200)
plt.subplot(336)
plt.imshow(contour , cmap="gray")
plt.title("detection des contour")
plt.axis("off")

plt.show()
```

```python
#exmple sur des image
contour =cv2.Canny(B , 100 , 200 )
plt.subplot(336)
plt.imshow(contour , cmap="gray")
plt.title("detection des contour")
plt.axis("off")

plt.show()
```
